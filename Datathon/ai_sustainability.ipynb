```python
exploded_df = ai_df.assign(Task=ai_df['Task'].str.split(',')).explode('Task')
exploded_df['Task'] = exploded_df['Task'].str.strip()  # Remove leading/trailing spaces
task_avg_df = exploded_df.groupby('Task')[['Training power (Watts) ', 'Training energy (kWh) ', 'Carbon Emissions from training (lb)']].mean().reset_index()
```

```python
metrics = ['Training power (Watts) ', 'Training energy (kWh) ', 'Carbon Emissions from training (lb)']

for metric in metrics:
    plt.figure(figsize=(10, 6))
    plt.bar(task_avg_df['Task'], task_avg_df[metric], edgecolor='black')
    plt.xlabel('Task')
    plt.ylabel(f'Average {metric}')
    plt.title(f'Average {metric} by Task')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```
